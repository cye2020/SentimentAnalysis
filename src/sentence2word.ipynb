```python
import pandas as pd
import json
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

nltk.download('punkt')                      
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
data = pd.read_csv('../Data/renual_data.csv')
data
```

```python
json_data = data.to_json(orient="records", date_format="iso")
js = json.loads(json_data)
js
```

```python
extract_specific_tags = ['JJ', 'JJR', 'JJS', 'NN', 'NNS', 'NNP', 'NNPS', 'RB', 'RBR', 'RBS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'] # Noun, Verb POS tags
stop_words = set(stopwords.words('english'))
```

```python
from datetime import datetime
```

```python
stop_words.update({"n't", "'m", "'s"})
```

```python
for row in js:
    # text -> words
    sentence = row['Text']
    pos_tokens = nltk.pos_tag(word_tokenize(sentence.lower()))
    tokenized_sentence = [
        token[0][1:] if token[0].startswith("'") else token[0]  # 'Î°ú ÏãúÏûëÌïòÎäî Í≤ΩÏö∞ Ï†úÍ±∞
        for token in pos_tokens
        if token[1] in extract_specific_tags and token[0] not in stop_words
    ]
    row['Words'] = tokenized_sentence
    
    # hashtags -> list of hashtags
    row['Hashtags'] = row['Hashtags'].split(' ')
```

```python
with open('../Data/data.json', 'w') as file:
    json.dump(js, file, indent=4)

print("JSON ÌååÏùºÏù¥ Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§: data.json")
```
