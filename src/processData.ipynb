```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

nltk.download('punkt')                      
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
extract_specific_tags = ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'] # Noun, Verb POS tags
stop_words = set(stopwords.words('english'))
```

```python
tokenized_sentences = []
for sentence in sentences:
    pos_tokens = nltk.pos_tag(word_tokenize(sentence.lower()))
    tokenized_sentence = [
        token for token in pos_tokens
        if token[1] in extract_specific_tags and not token[0] in stop_words
    ]
    tokenized_sentences.extend(tokenized_sentence)

len(tokenized_sentences), len(set(tokenized_sentences))
```

```python
words = [s[0] for s in tokenized_sentences]
print(*words, sep=", ")
```
